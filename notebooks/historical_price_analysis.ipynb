# AAPL

```python
AAPL_historical_data =pd.read_csv('C:/Users/nuhamin/Documents/kifiya/week 1/' \
'Stock-Market-Price-Prediction-with-Sentiment-Analysis/data/historical_data/AAPL_historical_data.csv')

AAPL_historical_data['Date'] = pd.to_datetime(AAPL_historical_data['Date'])
AAPL_historical_data = AAPL_historical_data[
    (AAPL_historical_data['Date'] >= '2020-01-01') &
    (AAPL_historical_data['Date'] <= '2020-12-31')]

AAPL_historical_data = AAPL_historical_data.reset_index(drop=True)
AAPL_historical_data.head()

stock_analyser = StockAnalyser(AAPL_historical_data)
ticker = 'AAPL'

#calculate technical indicators
data_with_indicators = stock_analyser.calculate_technical_indicators(stock_analyser.data)

#plot
fig = stock_analyser.plot_stock_data(data_with_indicators, ticker='AAPL', indicators=True, volume=True)

#display plot
display(fig)

#define plot name
plot_name = f'{ticker} - Historical Stock Price'

#save plot
stock_analyser.save_plot(fig=fig, plot_folder=plot_folder, plot_name=plot_name)

#save df locally
stock_analyser.save_dataframe(df=data_with_indicators,
                              df_folder=df_folder,
                              df_name=f'{ticker}_hist_data')
```
